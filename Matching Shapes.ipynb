```python
import numpy as np
import cv2

image = cv2.imread("E:\PVT\Academic\TO watch\SaleWebDesign.com-Master-Computer-Vision-Python-Deep\shape1.png")
original_image = image.copy()
cv2.imshow("original", original_image)
cv2.waitKey(0)

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray, 127, 255, 1)

contours, hierarchy = cv2.findContours(thresh.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
cv2.drawContours(image, contours, -1, (0,255,0), 3)
cv2.imshow("all contours in original", image)
cv2.waitKey(0) 

for cnt in contours:
    approx = cv2.approxPolyDP(cnt, 0.01 * cv2.arcLength(cnt, True), True)
    
    if len(approx) == 3:
        shape = "triangle"
        cv2.drawContours(image, [cnt], 0, (0,255,0), -1)
        M = cv2.moments(cnt)
        cx = int(M['m10'] / M['m00'])
        cy = int(M['m01'] / M['m00'])
        cv2.putText(image, shape, (cx-50, cy), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,0), 1)
        
    elif len(approx) == 4:
        x,y,w,h = cv2.boundingRect(cnt)
        M = cv2.moments(cnt)
        cx = int(M['m10'] / M['m00'])
        cy = int(M['m01'] / M['m00'])
        
        if abs(w-h)  <= 3:
            shape = "square"
            # find contour center to place text at the center
            cv2.drawContours(image, [cnt], 0, (0,125,255), -1)
            cv2.putText(image, shape, (cx-50, cy), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,0), 1)
            
        else:
            shape = 'rectangle'
            cv2.drawContours(image, [cnt], 0, (0,125,255), -1)
            cv2.putText(image, shape, (cx-50, cy), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,0), 1)
          
        
    elif len(approx) == 10:
        shape = "start"
        cv2.drawContours(image, [cnt], 0, (0,255,0), -1)
        M = cv2.moments(cnt)
        cx = int(M['m10'] / M['m00'])
        cy = int(M['m01'] / M['m00'])
        cv2.putText(image, shape, (cx-50, cy), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,0), 1)

    elif len(approx) >= 10:
        shape = "Circle"
        cv2.drawContours(image, [cnt], 0, (0,255,0), -1)
        M = cv2.moments(cnt)
        cx = int(M['m10'] / M['m00'])
        cy = int(M['m01'] / M['m00'])
        cv2.putText(image, shape, (cx-50, cy), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,0), 1)



    cv2.imshow("Shape", image)
    cv2.waitKey(0)
cv2.destroyAllWindows()
```
